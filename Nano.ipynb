```
from ultralytics import YOLO

# Charger un mod√®le pr√©-entra√Æn√© (par exemple, yolov8n)
model = YOLO('yolov8n.pt')
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
data_yaml = """
train: /content/drive/My Drive/yolo_images/train/images
val: /content/drive/My Drive/yolo_images/val/images
test: /content/drive/My Drive/yolo_images/test/images

nc: 3  # nombre de classes

names:
  - 'cubic'
  - 'pseudo-cubic'
  - 'ball'
"""

file_path = '/content/drive/My Drive/data.yaml'

try:
    with open(file_path, 'w') as f:
        f.write(data_yaml)
    print(f"File '{file_path}' written successfully.")
except Exception as e:
    print(f"Error writing file '{file_path}': {e}")
```

```
# Entra√Æner le mod√®le avec vos donn√©es
results = model.train(data='/content/drive/My Drive/data.yaml', epochs=30, imgsz=640, batch=16)
```

```
import matplotlib.pyplot as plt

results = model.predict(source='/content/drive/My Drive/yolo_images/val/images/369.jpg')
for result in results :
  annotated_image = result.plot(show=False)

  # Afficher l'image annot√©e avec Matplotlib
  plt.imshow(annotated_image)
  plt.axis('off')  # Masquer les axes
  plt.show()

  # Initialiser un dictionnaire pour compter les objets par classe
  class_counts = {}

  # Parcourir chaque bo√Æte de d√©limitation dans les r√©sultats
  for box in result.boxes:
      class_id = int(box.cls.item())  # Convertir le tensor en entier pour obtenir l'indice de classe
      class_name = result.names[class_id]  # Obtenir le nom de la classe √† partir de l'indice
      if class_name in class_counts:
          class_counts[class_name] += 1
      else:
          class_counts[class_name] = 1

  # Afficher le nombre d'objets d√©tect√©s par classe
  for class_name, count in class_counts.items():
      print(f"Classe '{class_name}': {count} objets d√©tect√©s")
```
